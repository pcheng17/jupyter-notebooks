```python
import pandas as pd
import json
import csv
import os
```

```python
with open('./data/alltime.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame.from_dict(data)
df['date'] = pd.to_datetime(df['date'])
df['yearmonth'] = df['date'].dt.to_period('M')
df_grouped = df.groupby(['yearmonth', 'artist'])['duration'].sum().reset_index()
df_grouped['date'] = df_grouped['yearmonth'].dt.to_timestamp()
df_grouped.drop('yearmonth', axis=1, inplace=True)


df_grouped = df_grouped.rename(columns={'artist': 'name', 'duration': 'value'})
df_grouped['category'] = 'kpop'
df_grouped = df_grouped[['date', 'name', 'category', 'value']]

df_grouped.sort_values(['name', 'date'], inplace=True)
df_grouped['accumulated'] = df_grouped.groupby('name')['value'].cumsum()

df_grouped.drop('value', axis=1, inplace=True)
df_grouped = df_grouped.rename(columns={'accumulated': 'value'})
df_grouped.sort_values('date', inplace=True)


df_grouped[df_grouped['name'] == 'Park Kyung']
```

```python
df_grouped.to_csv('alltime-grouped-artist.csv', index=False)
```
